```python
import matplotlib.pyplot as plt
import random as r
```

```python
dice = []
trials = 100

for i in range(trials):
  dice.append(r.randint(1,6))

min = trials
max = 0

for i in range(1,7):
  result = dice.count(i)
  print(f'{i} - {result} times')
  if(min>result):
    min = result
  if(max<result):
    max = result

print(f'Differnce {max - min}')
```

```python
plt.bar(list(range(1,7)), [dice.count(i) for i in range(1,7)], color = 'pink')
plt.show()
```
